```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\Other\OneDrive\DataScience\ASGN\football_mini.csv")   # data reading
```

```python
data.describe()   # data describing
```

```python
columns = ['Crossing', 'Finishing', 'ShortPassing', 'Dribbling']   # columns that are include in the heatmap
```

```python
def heatmap_top10(data, columns, save_path=None):   # defining the heatmap data
    top10_data = data[columns].head(10)   # selecting the top 10 rows from the dat
    plt.figure(figsize=(10, 8))   # Selecting the figure size
    correlation_matrix = top10_data.corr()   # Computing the correlation matrix for top 10 data
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")   # Creating a heatmap of correlation matrix with color annotations
    plt.title('Heatmap of Correlations (Top 10)', fontsize=16)   # Adding a title to the heatmap
    plt.xlabel('Attributes', fontsize=12)   # labeling the X-axis with attributes and font size
    plt.ylabel('Attributes', fontsize=12)   # labeling the Y-axis with attributes and font size
    if save_path:
        plt.savefig(save_path)   # path saving
    plt.show()   # displaying the plot
    
heatmap_top10(data, columns=columns, save_path='heatmap_top10.png')   # Calling the function with data and column list
```

```python
x_column = 'Age'       # Column for the x-axis
y_column = 'Wage'      # Column for the y-axis
```

```python
def scatter_plot_top10(data, x_column, y_column, save_path=None):   # defining the scatter plot data
    top10_data = data.head(10)   # selecting the top 10 rows from the data
    plt.figure(figsize=(8, 6))   # Selecting the figure size
    sns.scatterplot(data=top10_data, x=x_column, y=y_column)   # Creating a scatter plot using specified columns for x and y axis 
    plt.title(f'Scatter Plot of {x_column} vs {y_column} (Top 10)', fontsize=16)   # Adding a title to the plot 
    plt.xlabel(x_column, fontsize=12)   # labeling the X-axis with specified column name and font size
    plt.ylabel(y_column, fontsize=12)   # labeling the Y-axis with specified column name and font size
    if save_path:
        plt.savefig(save_path)   # path saving
    plt.show()   # displaying the plot

scatter_plot_top10(data, x_column, y_column, save_path='scatter_top10.png')# Calling the function with data and specified columns for x and y axis
```

```python
def bar_plot_top10(data, x_column, y_column, save_path=None):   # defining the data,x-col,y-col,saving path
    top10_data = data.head(10)   # selecting the top 10 rows from the data
    plt.figure(figsize=(12, 8))   # Selecting the figure size
    sns.barplot(data=top10_data, x=x_column, y=y_column)   # Creating a bar plot using specified columns for x and y axis
    plt.title(f'Bar Plot of {y_column} by {x_column} (Top 10)', fontsize=16)   # Adding a title to the plot 
    plt.xlabel(x_column, fontsize=12)   # labeling the X-axis with specified column name and font size
    plt.ylabel(y_column, fontsize=12)   # labeling the Y-axis with specified column name and font size
    if save_path:
        plt.savefig(save_path)   # path saving
    plt.show()   # displaying the plot
    
bar_plot_top10(data, x_column, y_column, save_path='Barplot_top10.png') # Calling the function with the data and specified columns for x and y axis
```

```python
column = 'Position'   # Defining the column for pie chart
```

```python
def pie_chart_top10(data, column, save_path=None):   # defining the data for pie chart
    top10_data = data.head(10)   # selecting the top 10 rows from the data
    plt.figure(figsize=(8, 8))   # Selecting the figure size
    top10_data[column].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, cmap='viridis', labels=top10_data[column].value_counts().index) 
    # autopct='%1.1f%%' displays the percentage values on each slice of pie chart
    # startangle=140 rotates the start angle of pie chart for better visualization
    # cmap='viridis' applies a colormap for a visually appealing color scheme
    # labels=top10_data[column].value_counts().index uses the index of value counts as labels
    plt.ylabel('')   # Removing the y-axis label for a cleaner pie chart appearance
    plt.title(f'Distribution of {column} (Top 10)', fontsize=16)   # adding a title 
    if save_path:
        plt.savefig(save_path)   # path saving
    plt.show()   # displaying the plot

pie_chart_top10(data, column, save_path='Piechart_Top10.png')   # Calling the function with data, specified column, and save path
```
